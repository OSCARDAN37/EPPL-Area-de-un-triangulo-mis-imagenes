# Formula para calcular el area
$A = (b \cdot h) / 2$
---
## Datos de entrada
* $h ?$
* $b ?$

```python
# Calcular area de un circulo
# variables h, b, A
# formula A = (h * b)/2
# Capturar el valor de A
import math
print("--- Calcular el área de un triángulo ---")
h = float(input("Ingrese el valor de la altura (h): "))
b = float(input("Ingrese el valor de la base (b): "))
A = (h * b) / 2
print("---------------------------------------")
print("El área del triángulo es:", A)
print("Oscar Flores NL = 19")
print("---------------------------------------")
```
